```python
import matplotlib.pyplot as plt
import scipy as sp
```

The RSA generator works differently than most of the other ones, in that it requires an empty image to be passed as an argument.  The RSA function will then add spheres of the specified size to this received image until the total volume fraction of the image reaches the specified value.  This workflow was used so that increasingly smaller spheres could be added to the image to create a polydisperse packing as shown below:

```python
im = sp.zeros([200, 200])
im = ps.generators.RSA(im=im, radius=15)
im = ps.generators.RSA(im=im, radius=10)
im = ps.generators.RSA(im=im, radius=5)
im = ps.generators.RSA(im=im, radius=3)
```

```python
fig = plt.figure(figsize=[12, 12])
plt.imshow(im)
```
